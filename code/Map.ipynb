```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

pd.options.display.max_columns=None
plt.style.use('seaborn-whitegrid')
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['legend.loc'] = 'upper right'
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 12

import warnings
warnings.filterwarnings('ignore')
```

```python
#Se carga el archivo 'estaciones.csv'

estaciones = pd.read_csv(r'estaciones.csv').drop(['indsinop', 'latitud', 'longitud'], axis=1)
estaciones
```

```python
gdf = gpd.read_file('Estaciones_Completas.shp')
gdf['longitud'] = gdf['geometry'].x
gdf['latitud'] = gdf['geometry'].y
gdf
```

```python
indicativo = list(estaciones['indicativo'])
coordenadas = gdf.query('INDICATIVO == @indicativo')[['INDICATIVO', 'longitud', 'latitud']]\
                    .reset_index(drop=True)\
                    .rename(columns={'INDICATIVO': 'indicativo'})

coordenadas_estaciones = estaciones.merge(coordenadas, how='left', on='indicativo')
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='2661', 'latitud'] = 42.58833127219514
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='2661', 'longitud'] = -5.651114881872618
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='7031', 'latitud'] = 37.788888121737074
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='7031', 'longitud'] = -0.8033320760122636
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='8416', 'latitud'] = 39.48055317460837
coordenadas_estaciones.loc[coordenadas_estaciones['indicativo']=='8416', 'longitud'] = -0.3663889192770184
coordenadas_estaciones
```

```python
# Se carga el mapa de las CCAA

ccaa = gpd.read_file('CCAA_GEO_ETRS89.shp')

# Se quitan Canarias, Ceuta y Melilla

ccaa = ccaa.query('cod_CCAA != ["CA05", "CA18", "CA19"]')

# Se grafica el mapa de color blanco con los contornos negros

ax = ccaa.plot(color='white', edgecolor='black', figsize=(20,20))

# Se marcan las estaciones meteorológicas con puntos azules

gdf_spain_blue = gpd.GeoDataFrame(coordenadas_estaciones,
                                  geometry=gpd.points_from_xy(coordenadas_estaciones['longitud'],
                                                              coordenadas_estaciones['latitud']))

gdf_spain_blue.plot(ax=ax, color='blue', markersize=40)

# Se añaden los nombres de las estaciones meteorológicas al lado de los puntos azules

for x, y, label in zip(gdf_spain_blue.geometry.x, gdf_spain_blue.geometry.y, gdf_spain_blue['nombre']):
    ax.annotate(label, xy=(x, y), xytext=(5,3), textcoords='offset points')

# Se visualiza el gráfico

plt.tight_layout()
plt.axis('off')
plt.show()
```
