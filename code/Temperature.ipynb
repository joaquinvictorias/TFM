```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np

pd.options.display.max_columns=None

import warnings
warnings.filterwarnings('ignore')
```

```python
# Obtenemos información de la estación seleccionada durante el periodo deseado
url = f'https://opendata.aemet.es/opendata/api/valores/climatologicos/diarios/datos/fechaini/2020-01-01T00:00:00UTC/fechafin/2022-12-31T23:59:59UTC/estacion/1484C'
url
```

```python
# Importamos la librería requests
import requests

# Cargamos la api key
querystring = {'api_key': '...'}

# Realizamos el request
r = requests.get(url, params=querystring, verify=False)
r_json = r.json()
r
```

```python
# Exploramos el archivo JSON
r.json()
```

```python
# Obtenemos el link del que descargaremos los datos
data_url = r.json()['datos']
r_data = requests.get(data_url, params=querystring, verify=False)

# Vemos el contenido
raw_data = r_data.json()
dataset = pd.DataFrame(raw_data)
dataset['fecha'] = pd.to_datetime(dataset['fecha'])
dataset
```
